```python
import pandas

#create dataframe

dataframe= pandas.read_csv('bank.csv')
```

```python
dataframe.head()
```

```python
dataframe.isnull().sum()
```

```python
dataframe['Self_Employed'].fillna('unknown',inplace=True)

avg= dataframe['LoanAmount'].mean()

dataframe.fillna('LoanAmount',inplace=True)

dataframe.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.style.use('seaborn-v0_8')

sns.barplot(x='Property_Area',y='ApplicantIncome',data= dataframe ,palette='magma')
plt.title('Property_Area against ApplicationIncome')
```

```python

sns.barplot(x='LoanAmount',y='Married',data= dataframe ,palette='Greys')
plt.title('Married against LoanAmount ')
```
