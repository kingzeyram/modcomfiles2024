```python
import pandas as pd
data=pd.read_csv('../DataScience/bank.csv')
data.head()
```

```python
#clean data up
data.isnull().sum()
```

```python
#delete rows with null values
data.dropna(inplace=True)
```

```python
#check the cleaned up data
data.isnull().sum()
```

```python
#split our dataframe into columns we intend to use for clustering

data.head()

array=data.values

array
```

```python
x = array[:,5:8]
x
```

```python
#feed data into model




from sklearn.cluster import KMeans

model=KMeans(n_clusters=10,random_state=42)

model.fit(x)
```

```python
#find center points/ centroids

#see the 10 clusters
means= model.cluster_centers_
clusters=pd.DataFrame(means,columns=['ApplicantIncome','CoapplicantIncome ','LoanAmount'])

clusters
```

```python
#view individual clusters

data['clusterno']=model.labels_


cluster1=data[data['clusterno']==1]
cluster1
```
