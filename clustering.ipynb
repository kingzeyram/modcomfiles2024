```python
# Unsupervised Ml
#Clustering. Train our model Without testing 

import pandas as pd
data=pd.read_csv('../DataScience/Mall_Customers.csv')
data.head(4)
```

```python
#clean data up
data.isnull().sum()
#delete rows with null values
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
#split our dataframe into columns we intend to use for clustering

data.head()

array=data.values

array

#array=data[:, 2:5]
```

```python
x = array[:, 2:5]
x
```

```python
#feed data into model


#sklearn has 3 algorithims
#k means
#k prototype
#k modes

from sklearn.cluster import KMeans

model=KMeans(n_clusters=12,random_state=42)

model.fit(x)

```

```python
#find center points/ centroids

#see the 10 clusters
means= model.cluster_centers_
clusters=pd.DataFrame(means,columns=['Age','Income','SpendindScore'])

clusters
```

```python
#view individual clusters

data['clusterno']=model.labels_

cluster1=data[data['clusterno']==1]
cluster1
```

```python
data['clusterno']=model.labels_

cluster2=data[data['clusterno']==2]
cluster2
```
